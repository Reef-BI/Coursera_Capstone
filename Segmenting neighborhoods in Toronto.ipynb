### Question 1. Wrangling data and analysis preparation.

I will use BeautifullSoup to parse the table from the web: https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M  and transform it to a dataframe

Lets import all libraries we require to assign this task:

```python
!conda install -c anaconda beautifulsoup4
!conda install -c conda-forge folium=0.5.0 --yes
import folium
import urllib
import bs4 as bs
from bs4 import BeautifulSoup
import requests
print("All libraries are imported!")
```

```python
url = requests.get("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")
soup = bs.BeautifulSoup(url.content, 'html.parser')
print(soup.prettify())
```

```python
postal_codes_table = soup.find("table", {'class':'wikitable sortable'})
postal_codes_table
```

```python
tr_rows = postal_codes_table.find_all('tr')
tr_rows
```

```python
tr_rows = tr_rows[1:]
```

```python
tr_rows[0].text.split('\n')
```

```python
import pandas as pd

columns_new_df = ['PostCode','Borough','Neighboorhood']

df_neighborhood = pd.DataFrame(columns = columns_new_df)

for row in tr_rows:
    row = row.text.split('\n')
    
    
    postcode = row[1]
    
    
    borough = neighborhood = row[2]
       
        
    if row[3] != 'Not assigned':
        neighborhood = row[3]
        
        
    df_neighborhood = df_neighborhood.append({'PostCode': postcode,
                                      'Borough': borough,
                                      'Neighboorhood': neighborhood}, ignore_index = True)

df_neighborhood.head(10)
```

```python
not_assigned = df_neighborhood[df_neighborhood['Borough'] == "Not assigned"].index
```

```python
df_neighborhood.drop(not_assigned , inplace = True)
```

```python
df_neighborhood.shape
```

```python
df_neighborhood.loc[df_neighborhood['Neighboorhood'] == 'Not assigned' , 'Neighboorhood'] = df_neighborhood["Borough"]
```

```python
df_neighborhood.head(20)
```

```python
splitted_df = df_neighborhood.groupby(['PostCode','Borough'], sort=False).agg(', '.join).reset_index()
```

```python
splitted_df.head(20)
```

```python
splitted_df.shape
```

```python
def getNearbyVenues(names, latitudes, longitudes, radius =500, LIMIT = 100):
    
    venues_list=[]   
    for name, lat, lng in zip(names, latitudes, longitudes):        
        print(name)
    
        webdata = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={} &ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID,
            CLIENT_SECRET,
            VERSION,
            lat,
            lng,
            radius,
            LIMIT)

        results = requests.get(webdata).json()["response"]['groups'][0]['items']

        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])
        
    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
     
    nearby_venues.columns = ['Neighborhood', 
                    'Neighborhood Latitude',
                    'Neighborhood Longitude', 
                    'Venue',
                    'Venue Latitude',                
                    'Venue Longitude',
                    'Venue Category']
        
    return(nearby_venues)
```

```python
toronto_venues = getNearbyVenues(names = postcds_and_crdnts_df['Neighborhood'], 
                                latitudes = postcds_and_crdnts_df['Latitude'],
                                longitudes = postcds_and_crdnts_df['Longitude']
                                )
```

### one hot encoding and returning the neighborhood column back

```python
toronto_onehot = pd.get_dummies(toronto_venues[['Venue Category']], prefix="", prefix_sep ="")

toronto_onehot['Neighborhood'] = toronto_venues['Neighborhood']

#making neighborhood first column

changed_columns = [toronto_onehot.columns[-1]] + list(toronto_onehot.columns[:-1])
toronto_onehot = toronto_onehot[changed_columns]

toronto_onehot.head()
```

```python
toronto_onehot.shape
```

```python
toronto_grouped = toronto_onehot.groupby('Neighborhood').mean().reset_index()
```

```python
toronto_grouped.shape
```

We will print each neighborhood along with the top 10 most common venues

```python
top_10_venues = 10
```

```python
for hood in toronto_grouped['Neighborhood']: 
    print("----"+hood+"----") 
    temp = toronto_grouped[toronto_grouped['Neighborhood'] == hood].T.reset_index()
    temp.columns = ['venue','freq'] 
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(top_10_venues)) 
    print('\n')
```

```python
def return_most_common_venues(row, top_10_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    return row_categories_sorted.index.values[0:top_10_venues]
```

```python
import numpy as np

top_10_venues = 10

indicators = ['st', 'nd', 'rd']

columns = ['Neighborhood'] 

for ind in np.arange(top_10_venues):
    try: 
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except: 
        columns.append('{}th Most Common Venue'.format(ind+1))
    
neighborhoods_venues_sorted = pd.DataFrame(columns=columns)

neighborhoods_venues_sorted['Neighborhood'] = toronto_grouped['Neighborhood']

for ind in np.arange(toronto_grouped.shape[0]):
    neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(toronto_grouped.iloc[ind, :], top_10_venues) 
    
neighborhoods_venues_sorted.head()
```

#### Clustering with k-means

```python
import numpy as np
# we will run 5 clusters 
kclusters = 5
```

```python
toronto_grouped_clusters = toronto_grouped.drop("Neighborhood", 1)
```

```python
kmeans = KMeans(n_clusters=kclusters, random_state = 0).fit(toronto_grouped_clusters)
kmeans.labels_[0:40]
```

```python
neighborhoods_venues_sorted.insert(0, 'Cluster_Labels', kmeans.labels_)
```

We will create new DataFrame and join top venues per Neighborhood

```python
toronto_merged = postcds_and_crdnts_df

toronto_merged = toronto_merged.join(neighborhoods_venues_sorted.set_index('Neighborhood'), on='Neighborhood')

toronto_merged=toronto_merged.dropna() # we need to remove NA values and convert it to int to be able display everything on map

toronto_merged['Cluster_Labels'] = toronto_merged.Cluster_Labels.astype('int')
```

```python
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

markers_colors = []
for lat, lon, poi, cluster in zip(toronto_merged['Latitude'], toronto_merged['Longitude'], toronto_merged['Neighborhood'],
toronto_merged['Cluster_Labels']):
    label = folium.Popup(str(poi)+' Cluster '+ str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label, 
        color=rainbow[cluster-1],
        fill=True, 
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
    
map_clusters
```
